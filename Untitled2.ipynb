```python
drop_cols = [
    's.no',
    'about_item',
    'best_sellers_rank',
    'brand_page_url',
    'customer_review_summary',
    'default_variant/0',
    'default_variant/1',
    'default_variant/2',
    'delivery_date',
    'fastest_delivery_date',
    'manufacturer',
    'model_number',
    'product_description',
    'product_url',
    'rating_distribution/1star',
    'rating_distribution/2star',
    'rating_distributionion/3star',
    'rating_distribution/4star',
    'rating_distribution/5star',
    'recent_purchases',
    'scrape_time',
    'seller_page_url',
    'title',
    'all_images',
    'rank_1'
]

products_clean = products.drop(columns=drop_cols, errors='ignore')

```

```python
products_clean.shape
```

```python
products_clean.columns
```

```python
products_clean.info()
```

```python
products_clean.isnull().sum()
```

```python
num_cols = ['list_price', 'price_value', 'rating_count', 'rating_stars']

for col in num_cols:
    products_clean[col] = (
        products_clean[col]
        .astype(str)
        .str.replace('[^0-9.]', '', regex=True)
    )

products_clean[num_cols] = products_clean[num_cols].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
products_clean[num_cols] = products_clean[num_cols].fillna(
    products_clean[num_cols].median()
)
```

```python
cat_cols = ['availability', 'breadcrumbs', 'seller_name']

for col in cat_cols:
    products_clean[col] = products_clean[col].fillna(
        products_clean[col].mode()[0]
    )
```

```python
products_clean.info()
products_clean.isnull().sum()
```

```python
reviews = pd.read_csv("reviews.csv")
reviews
```

```python
reviews.columns
reviews.info()
```

```python
drop_cols = [
    's.no',
    'images/0', 'images/1', 'images/2', 'images/3',
    'images/4', 'images/5', 'images/6', 'images/7',
    'videos/0',
    'reviewID',
    'reviewMetadata',
    'reviewPosition',
    'reviewURL',
    'reviewTitle'
]

reviews_clean = reviews.drop(columns=drop_cols, errors='ignore')
```

```python
reviews_clean = reviews_clean[
    [
        'productASIN',
        'rating',
        'reviewText',
        'verifiedPurchase',
        'cleaned_review_text',
        'sentiment_score'
    ]
]
```

```python
reviews_clean['rating'] = pd.to_numeric(
    reviews_clean['rating'], errors='coerce'
)
```

```python
reviews_clean['verifiedPurchase'] = (
    reviews_clean['verifiedPurchase']
    .astype(int)
)
```

```python
reviews_clean['rating'] = reviews_clean['rating'].fillna(
    reviews_clean['rating'].median()
)

reviews_clean['sentiment_score'] = reviews_clean['sentiment_score'].fillna(
    reviews_clean['sentiment_score'].median()
)
```

```python
reviews_clean['reviewText'] = reviews_clean['reviewText'].fillna('')
reviews_clean['cleaned_review_text'] = reviews_clean['cleaned_review_text'].fillna('')
```

```python
reviews_clean['review_length'] = reviews_clean['cleaned_review_text'].apply(len)
```

```python
reviews_clean.info()
```

```python
reviews_clean.isnull().sum()
```

```python
reviews_clean['is_negative'] = (reviews_clean['rating'] <= 2).astype(int)
```

```python
review_agg = (
    reviews_clean
    .groupby('productASIN')
    .agg(
        avg_review_rating=('rating', 'mean'),
        rating_variance=('rating', 'var'),
        review_volume=('rating', 'count'),
        negative_review_ratio=('is_negative', 'mean'),
        avg_sentiment=('sentiment_score', 'mean'),
        avg_review_length=('review_length', 'mean'),
        verified_ratio=('verifiedPurchase', 'mean')
    )
    .reset_index()
)
```

```python
review_agg['rating_variance'] = review_agg['rating_variance'].fillna(0)
```

```python
review_agg.info()
review_agg.head()
review_agg.isnull().sum()
```

```python
review_agg = review_agg.rename(columns={'productASIN': 'asin'})
```

```python
df = products_clean.merge(
    review_agg,
    on='asin',
    how='left'
)
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
review_cols = [
    'avg_review_rating',
    'rating_variance',
    'review_volume',
    'negative_review_ratio',
    'avg_sentiment',
    'avg_review_length',
    'verified_ratio'
]

df[review_cols] = df[review_cols].fillna({
    'avg_review_rating': 0,
    'rating_variance': 0,
    'review_volume': 0,
    'negative_review_ratio': 0,
    'avg_sentiment': 0,
    'avg_review_length': 0,
    'verified_ratio': 0
})
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['satisfaction_score'] = (
    0.5 * df['avg_review_rating']
    + 0.3 * df['avg_sentiment']
    - 0.2 * df['negative_review_ratio']
)
```

```python
df['satisfaction_score'].describe()
```

```python
low_cut = df['satisfaction_score'].quantile(0.75)
high_cut = df['satisfaction_score'].quantile(0.25)
```

```python
def assign_risk(score):
    if score >= low_cut:
        return 0   # Low Risk
    elif score <= high_cut:
        return 2   # High Risk
    else:
        return 1   # Medium Risk

df['risk_level'] = df['satisfaction_score'].apply(assign_risk)
```

```python
df['risk_level'].value_counts(normalize=True)
```

```python
features = [
    'availability',
    'brand_name',
    'breadcrumbs',
    'list_price',
    'price_value',
    'rating_count',
    'rating_stars',
    'seller_name',

    # review-based features
    'avg_review_rating',
    'rating_variance',
    'review_volume',
    'negative_review_ratio',
    'avg_sentiment',
    'avg_review_length'
]

X = df[features]
y = df['risk_level']
```

```python
cat_cols = X.select_dtypes(include='object').columns
num_cols = X.select_dtypes(exclude='object').columns
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    encoders[col] = le
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight='balanced'
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(model, "risk_model.pkl")
joblib.dump(encoders, "encoders.pkl")
joblib.dump(features, "feature_columns.pkl")
```

```python
new_raw = {
    "availability": "In Stock",
    "brand_name": "Apple",
    "breadcrumbs": "Electronics > Laptops",
    "list_price": 90000,
    "price_value": 87999,
    "rating_count": 950,
    "rating_stars": 4.6,
    "seller_name": "AppleStore",

    "avg_review_rating": 4.5,
    "rating_variance": 0.25,
    "review_volume": 820,
    "negative_review_ratio": 0.05,
    "avg_sentiment": 0.65,
    "avg_review_length": 180
}
```

```python
def safe_encode(value, encoder):
    if value in encoder.classes_:
        return encoder.transform([value])[0]
    else:
        return -1   # unseen category
```

```python
import pandas as pd

encoded_input = {}

for col in features:
    if col in encoders:   # categorical column
        encoded_input[col] = safe_encode(new_raw[col], encoders[col])
    else:                 # numerical column
        encoded_input[col] = new_raw[col]

X_unknown = pd.DataFrame([encoded_input])
```

```python
pred = model.predict(X_unknown)[0]
proba = model.predict_proba(X_unknown)[0]
```

```python
risk_map = {
    0: "Low Risk 🟢",
    1: "Medium Risk 🟡",
    2: "High Risk 🔴"
}

print("Predicted Risk:", risk_map[pred])
print("Probability Breakdown:")
print({
    "Low Risk": round(proba[0], 2),
    "Medium Risk": round(proba[1], 2),
    "High Risk": round(proba[2], 2),
})
```
